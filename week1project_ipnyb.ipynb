```python
# First step : Importing the important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Second step : Load the dataset
# (Download 'city_day.csv' from Kaggle Air Quality dataset and keep in same folder)
df = pd.read_csv("city_day.csv")
```

```python
# Third step : Explore the dataset
print("Shape of Dataset:", df.shape)
print("\nFirst 5 Rows:\n", df.head())

print("\n--- Dataset Info ---")
print(df.info())

print("\n--- Statistical Summary ---")
print(df.describe())

print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- First five rows ---")
print(df.head())

print("\n--- Last five rows ---")
print(df.tail())
```

```python
# Fourth step : Optional: Check AQI bucket distribution if available
if 'AQI_Bucket' in df.columns:
    print("\nAQI_Bucket Value Counts:\n", df['AQI_Bucket'].value_counts())

    # Plot AQI category distribution
    plt.figure(figsize=(7,4))
    sns.countplot(data=df, x="AQI_Bucket", palette="coolwarm", order=df["AQI_Bucket"].value_counts().index)
    plt.title("Distribution of AQI Categories")
    plt.xticks(rotation=30)
    plt.show()
```

```python
# Fifth step : Optional: Correlation Heatmap for pollutants
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlOrRd", fmt=".2f")
plt.title("Correlation Heatmap of Pollutants & AQI")
plt.show()
```
